## Read multiple data files using glob

```python
# import libraries
import pandas as pd
import glob
from os.path import isfile, join
```

```python
# glob finds all the pathnames matching a specified string
allFiles = glob.glob("./201*_PopularBabyNames.xlsx")
allFiles
```

```python
# Read multiple files
multiple_files = [pd.read_excel(file) for file in allFiles]
```

```python
# display multiple files
pd.concat(multiple_files, ignore_index = True)
```
