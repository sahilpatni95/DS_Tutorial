# Resize images

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from glob import glob
import cv2
```

```python
# reading all the images
images = []
for i in glob('../Data/Resize_images/*.jpeg'):
    images.append(plt.imread(i))
images = np.array(images)
# shape of original images 
images.shape
```

```python
# since each image is of different shape, we are only getting the number of images in the output
```

```python
# visualizing the original images
fig,ax = plt.subplots(nrows=1,ncols=4,figsize=(20,20))
for i in range(4):
    ax[i].imshow(images[i])
    ax[i].axis('off')
```

```python
# resizing all the images to same size
# defining the required shape
required_shape = (224,224,3)

resized_images = []
for i in range(images.shape[0]):
    # resizing the images to same size
    resized_images.append(cv2.resize(images[i], (224,224)))
resized_images = np.array(resized_images)
resized_images.shape
```

```python
# visualizing the original images
fig,ax = plt.subplots(nrows=1,ncols=4,figsize=(20,20))
for i in range(4):
    ax[i].imshow(resized_images[i])
    ax[i].axis('off')
```
