# Save Machine Learning Model using Pickle

```python
#import libraries
import pandas as pd
import numpy as np
import sklearn 
```

```python
#import dataset
data = pd.read_csv('../Data/data_cleaned.csv')
data.head()
```

## Build ML model

```python
# Define Dependent and independent variable
X = data.drop('Survived', axis = 1)
y = data['Survived']

# Split train and test data
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state = 142)

# import KNeighborsClassifier model 
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors = 3) 
  
# train model 
knn.fit(train_X, train_y) 
```

## Save the model

```python
import pickle

# Save the KNN model using pickle
saved_model = pickle.dumps(knn)

# Load the saved  KNN model
load_model = pickle.loads(saved_model) 

# Make predictions using the loaded pickled model 
load_model.predict(test_X) 
```
