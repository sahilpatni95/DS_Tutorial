```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Generate and display dataframe
np.random.seed(30)
df = pd.DataFrame(np.random.randn(10, 5), columns=list('ABCDE'))
df
```

```python
# Create heatmap using seaborn's style gradient
hm = sns.light_palette('green',as_cmap = True )
style = df.style.background_gradient(cmap = hm)
style
```

```python
df.style.set_caption("Pandas Styling")
```
