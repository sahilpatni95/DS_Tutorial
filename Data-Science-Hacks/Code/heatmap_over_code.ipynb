```
%%heat

import pandas as pd
import random
import math

df = pd.DataFrame({
    'A' : [random.randint(15,20) for i in range(1,1000) ],
    'B' : [random.randint(10,30) for i in range(1,1000) ]
})

def func(x):
    return math.sin(x.A**2) + math.sin(x.B**2) + math.tan(x.A**2)

res = df.apply(func, axis=1)
```
